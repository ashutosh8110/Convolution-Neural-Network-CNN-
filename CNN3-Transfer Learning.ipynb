```
from keras.preprocessing.image import ImageDataGenerator
train=ImageDataGenerator(preprocessing_function=preprocess_input)
val=ImageDataGenerator(preprocessing_function=preprocess_input)
```

```
train_gen=train.flow_from_directory("Images/",target_size=(224,224),batch_size=32,class_mode="categorical")
```

```
val_gen=val.flow_from_directory("val/",target_size=(224,224),batch_size=32,class_mode="categorical")
```

```
from keras.applications.resnet50 import ResNet50,preprocess_input,decode_predictions
```

```
model= ResNet50(weights="imagenet",input_shape=(224,224,3),include_top=False)
```

```
from keras.layers import *
from keras.models import Model
```

```
avl=GlobalAveragePooling2D()(model.output)
fc1=Dense(256,activation="relu")(avl)
d1=Dropout(0.5)(fc1)
fc2=Dense(4,activation="softmax")(d1)
```

```
model.output
```

```
complete_model=Model(inputs=model.input,output=fc2)
```

```
complete_model.summary()
```

```
for ix in range(len(complete_model.layers)):
  print(ix,complete_model.layers[ix].name)
```

```
for ix in range(174):
  complete_model.layers[ix].trainable=False
```

```
complete_model.summary()
```

```
from keras.optimizers import Adam
```

```
adam=Adam(lr=.003)
complete_model.compile(optimizer="adam",metrics=['accuracy'],loss="categorical_crossentropy")
```

```
history=complete_model.fit_generator(train_gen,epochs=4,steps_per_epoch=6,validation_data=val_gen,validation_steps=1)
```
